kalau masih ingat, saat membuat function, kita ada belajar tentang Scope variable. kita ada belajar best practice, function tidak mengubah data yang masuk ke dalam function tersebut, tapi mengembalikan data yang sudah selesai di proses saja. data awal jangan diusik. dengan konsep yang sama, function `rename` tidak mengubah data awal kita df_nba_clean. dia hanya *return* data baru dengan kolom yang sudah di rename.

untuk menyimpan hasil perubahan, kita perlu reassign atau masukin ke variable baru. ada beberapa opsi:
- reassign --- df_nba_clean = df_nba_clean.rename...
- inplace --- opsi inplace untuk mengubah data awal kita langsung, hasil sama seperti reassign, caranya adalah `df_nba_clean.rename(... inplace=True)`
- variable baru --- hasil df_nba_clean yang sudah kita rename column nya, kita masukin ke variable baru. cara ini paling aman kalau kita melakukan kesalahan, kita bisa balik ambil variable sebelumnya untuk ulang. cuman cara ini paling banyak variable

```python
# reassign
df_nba_clean = df_nba_clean.rename(columns={"game_id":"Game ID"})

# pakai inplace
df_nba_clean.rename(columns={"game_id":"Game ID"}, inplace=True)

# pakai variable baru
df_nba_clean_renamed = df_nba_clean.rename(columns={"game_id":"Game ID"})
```

pilih satu cara diatas dan run, kalau di run semua akan error karena kolom game_id sudah tidak ada, sudah berubah jadi Game ID di baris kedua
