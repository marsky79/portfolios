STOPWORDS adalah kumpulan kata unik yang biasanya kita tidak mau tampilkan di sebuah wordcloud, kita bisa tambahkan/ kurangi sendiri kata kata ke dalam STOPWORDS sebelum digunakan untuk generate word cloud kita

```python
stopword_kita = list(STOPWORDS) + ["said"]
stopword_kita = set(stopword_kita)
```

```python
wc_alice = WordCloud(
    background_color="white",
    max_words=2000,
    stopwords=stopword_kita
)

wc_alice = wc_alice.generate(novel_alice)

plt.imshow(wc_alice)
plt.axis("off")

plt.show()
```

tambahan secara visual:
- ubah ukuran word cloud
- menggunakan interpolation untuk menghaluskan edges gambar

```python
wc_alice = WordCloud(
    background_color="white",
    max_words=2000,
    stopwords=stopword_kita
)

wc_alice = wc_alice.generate(novel_alice)

# ubah ukuran
fig = plt.figure()
fig.set_figwidth(20)
fig.set_figheight(10)

plt.imshow(wc_alice,
           interpolation="bilinear" # membuat edges gambar lebih halus
           )
plt.axis("off")

plt.show()
```

# Menggunakan Mask

kita akan gunakan package PIL, module Image, untuk import file png ke python https://pillow.readthedocs.io/en/stable/reference/Image.html

```python
alice_mask = Image.open("alice_mask.png")

alice_mask
```

gambar di bahasa pemrograman, termasuk python, disimpan sebagai kumpulan angka.

Biasanya gambar hitam putih punya 1 layer, 1 set data dua dimensi numpy matrix, setiap angka di dalam numpy matrix ini value nya di atnara 0-255 untuk seberapa gelap warnanya.

```python
df_alice_mask = pd.DataFrame(np.array(alice_mask))
df_alice_mask.loc[100].value_counts()
```

karena gambar kita hitam putih dua warna saja, maka value nya hanya dua juga, yaitu 255 (putih) dan 0 (hitam)

untuk gambar yang grayscale, dimana hitamnya ada yang lebih gelap/ lebih terang akan ada value di antara 0-255

untuk gambar yang berwarna, biasanya ada 3 layer numpy array seperti ini, masing masing untuk warna red, green, blue.

```python
arr_alice_mask = np.array(alice_mask)
```

```python
wc_alice = WordCloud(
    background_color="white",
    max_words=2000,
    stopwords=stopword_kita,
    mask = arr_alice_mask
)

wc_alice = wc_alice.generate(novel_alice)

# ubah ukuran
fig = plt.figure()
fig.set_figwidth(20)
fig.set_figheight(10)

plt.imshow(wc_alice,
           interpolation="bilinear" # membuat edges gambar lebih halus
           )
plt.axis("off")

plt.show()
```

# Data Tabular
bagaimana kalau kita mau gunakan data bukan novel/ seminar/ presentasi tapi lebih data tabular biasa dan dijadikan wordcloud?

Misalkan, data imigrasi ke Canada kemarin. Kita bisa buat word cloud dimana, tergantung jumlah imigran yang dikirim ke Canada, nama negaranya akan muncul lebih besar di word cloudnya. tergantung frekuensi, tapi data kita bentuknya numerik.

misalnya... Indonesia 100, Malaysia 200, Singapore 80

bisakah data seperti ini jadi word cloud? tentu bisa, tapi kita perlu konversi dari data numerik 100 200 80 menjadi data text. misal membuat text dengan Indonesia Indonesia Indonesia Indonesia... sampai 100 kali, dan negara lain dengan jumlah frekuensinya.

lalu, misalkan datanya angkanya besar 100000, 2jt, dll, daripada membuat text yang sangat besar, kita bisa ubah skalanya juga.

contoh: dari Indonesia 100, Malaysia 200, Singapore 80 ==> Indonesia 10 Malaysia 20 Singapore 8 jadi hasil akhir text kita lebih kecil, dengan hasil wordcloud yang sama.

```python
# import dan rapikan data Canada kemarin

df_can = pd.read_excel("Canada.xlsx",
                        sheet_name="Canada by Citizenship",
                        skiprows=20,
                        skipfooter=2)

df_can = df_can.rename(columns={"OdName":"Country", "AreaName":"Continent", "RegName":"Region"})

df_can = df_can.drop(columns=["Type", "Coverage", "AREA", "REG", "DEV", "DevName"])

df_can = df_can.set_index("Country")

years = range(1980,2014)

df_can.head()
```

kita mau dapatkan total frekuensi berapa kali nama negaranya akan muncul di text sebelum kita jadikan wordcloud. Kita mau gunakan data total imigrasi dari negara tersebut ke Canada sebagai acuan. Kita perlu buat dulu total nya.

```python
total_imigran_per_negara = df_can[years].sum(axis=1)
total_imigran_per_negara
```

dari total ini kita buat text. perhatikan nilai nya lumayan besar:

```python
total_imigran_per_negara.sum()
```

kalau kita langsung konversi 1 to 1, per imigran nama negara muncul 1 kali, kita akan membuat text dengan isi lebih dari 6jt nama negara

kita kurangi saja skalanya, misalnya 1 kali muncul per 1000 imigran. ini artinya, negara yang total imigran ke Canada dibawah 1000 tidak akan kelihatan di word cloudnya, karena tidak ada di text.

karena mau untuk wordcloud, kita mau ubah jadi integer. tidak ada konsep stengah kata.

```python
skala = 1000

count_negara = (total_imigran_per_negara/skala).astype(int)
count_negara
```

```python
count_negara.sum()
```

```python
# convert dari count menjadi text

text_negara = ""

for country in count_negara.index:
    # print(country, count_negara[country])
    nama_negara = country
    freq = count_negara[country]

    # kita tempelkan nama negara jadi satu, misalnya "United States" menjadi "UnitedStates"
    nama_negara = nama_negara.replace(" ", "")
    # kita hapus juga punctuation di nama negaranya seperti ()'-,
    nama_negara = nama_negara.replace("(", "")
    nama_negara = nama_negara.replace(")", "")
    nama_negara = nama_negara.replace("'", "")
    nama_negara = nama_negara.replace("-", "")
    nama_negara = nama_negara.replace(",", "")

    text_negara = text_negara + (" " + nama_negara) * freq

text_negara
```

```python
# baru buat word cloud
wc_can_immi = WordCloud(
    background_color="white",
    max_words=195,
    collocations=False
)

wc_can_immi.generate(text_negara)

fig = plt.figure()
fig.set_figwidth(20)
fig.set_figheight(10)

plt.imshow(wc_can_immi)
plt.axis("off")
plt.show()
```
