```python
import pandas as pd
```

```python
# Sample dataset
data = pd.DataFrame({
    'device_name': ['DeviceA', 'DeviceB', 'DeviceC', 'DeviceD'],
    'issue_type': ['Battery drain', 'Screen flickering', 'Overheating', 'Wi-Fi drops'],
    'issue_frequency': [10, 5, 20, 7],
    'severity': ['Major', 'Minor', 'Critical', 'Moderate'],
    'solution_effectiveness': [4, 2, 5, 3],
    'user_ratings': [4.5, 3.8, 4.2, 3.9],
    'sales_data': [10000, 15000, 5000, 20000],
    'maintenance_frequency': [3, 1, 5, 2],
    'maintenance_cost': [20, 10, 30, 15],
    'warranty_claims': [2, 1, 3, 1],
    'warranty_coverage': ['Full', 'Partial', 'Full', 'Partial']
})
```

```python
data
```

```python
# Severity mapping
severity_mapping = {'Critical': 5, 'Major': 4, 'Moderate': 3, 'Minor': 2, 'Trivial': 1}
data['severity_score'] = data['severity'].map(severity_mapping)
```

```python
# Feasibility score calculation example
data['feasibility_score'] = (data['solution_effectiveness'] * data['user_ratings'] / (data['issue_frequency'] * data['severity_score']))
```

```python
# Display the dataframe with feasibility scores
print(data[['device_name', 'feasibility_score']])
```
